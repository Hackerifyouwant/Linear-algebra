```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
v = np.array([3, -2])
thetas = np.linspace(0, 2*np.pi, 100)
vecmags = np.zeros((thetas.shape[0], 2))
```

```python
for i in range(thetas.shape[0]):
    theta = thetas[i]
    A1 = np.array([[2 * math.cos(theta), -math.sin(theta)], [math.sin(theta),math.cos(theta)]])
    A2 = np.array([[math.cos(theta), -math.sin(theta)], [math.sin(theta),math.cos(theta)]])
    vecmags[i][0] = np.linalg.norm(A1 @ np.transpose(v))
    vecmags[i][1] = np.linalg.norm(A2 @ np.transpose(v))
```

```python
plt.plot(thetas, vecmags, 'o-')
plt.xlabel('Angle (rad.)')
plt.ylabel('Av agnitude')
plt.legend(['impure rotation', 'pure rotation'])
plt.show()
```
